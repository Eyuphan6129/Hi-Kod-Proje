```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mat2.csv")
df.head()
```

```python
df_cleaned = df.drop(columns = ["Unnamed: 0"])
df_cleaned.head()
```

```python
categorical_cols = df_cleaned.select_dtypes(include = ["object"]).columns

df_encoded = pd.get_dummies(df_cleaned, columns = categorical_cols, drop_first = True, dtype = "int")

df_encoded.head()
```

```python
X = df_encoded.drop(columns = ["G3"])
y = df_encoded["G3"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
models = {
    "Linear Regression" : LinearRegression(),
    "Decision Tree" : DecisionTreeRegressor(),
    "Random Forest" : RandomForestRegressor()
}
```

```python
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = y_pred
```

```python
performance = {}
for name, y_pred in results.items():
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    performance[name] = {"MAE" : mae, "RMSE" : rmse, "R2 Score" : r2}
```

```python
sorted_performance = sorted(performance.items(), key = lambda x : x[1]["R2 Score"], reverse = True)
```
